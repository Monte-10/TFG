```python
import pandas as pd

# Cargamos el archivo Excel en un dataframe de pandas, omitiendo las filas que no son parte de los datos
df = pd.read_excel('FoodCSV.xlsx', engine='openpyxl')

# Lista corregida de nombres de columna
df.columns = [
    'name', 'unit_weight', 'calories', 'protein', 'carbohydrates', 'sugar',
    'fiber', 'fat', 'saturated_fat', 'gluten_free', 'lactose_free',
    'pescetarian', 'vegetarian', 'vegan', 'contains_meat',
    'contains_vegetables', 'contains_fish_shellfish_canned_preserved',
    'cereal', 'pasta_or_rice', 'dairy_yogurt_cheese', 'fruit', 'nuts', 'legume',
    'sauce_or_condiment', 'deli_meat', 'bread_or_toast', 'egg',
    'special_drink_or_supplement', 'tuber', 'other'
]



# Verifica que los nombres de las columnas se hayan asignado correctamente
print("Nombres de las columnas después de la asignación:")
print(df.columns.tolist())

# Convertimos las columnas booleanas de '0' y '1' a valores booleanos
bool_columns = [
    'gluten_free', 'lactose_free', 'pescetarian', 'vegetarian', 'vegan', 'contains_meat',
    'contains_vegetables', 'contains_fish_shellfish_canned_preserved',
    'cereal', 'pasta_or_rice', 'dairy_yogurt_cheese', 'fruit', 'nuts', 'legume',
    'sauce_or_condiment', 'deli_meat', 'bread_or_toast', 'egg',
    'special_drink_or_supplement', 'tuber', 'other'
]

for column in bool_columns:
    df[column] = df[column].astype(bool)

# Verificamos las primeras filas del DataFrame
df.head()

# Guardamos df en un archivo CSV y lo guardamos en el directorio actual
df.to_csv('FoodCSV.csv', index=False)




```
